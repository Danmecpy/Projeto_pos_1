```python
import duckdb
import pandas as pd
```

```python
con = duckdb.connect(database='dados_duckdb.db', read_only=False)
```

```python
#ver todas as tabelas do banco de dados_duckdb.db
df_tables = con.execute("SHOW TABLES").fetchdf()
df_tables.head()

```

```python
df = con.execute('SELECT * FROM produtos').fetchdf()
```

```python
df2 = df.drop(columns=['id_categoria', 'id_fornecedor'] )
```

```python
con.execute("""
            DROP TABLE IF EXISTS dim_produtos""")
```

```python
con.execute("""CREATE TABLE IF NOT EXISTS dim_produtos (
id_produto BIGINT,
nm_produto VARCHAR,
vl_preco DECIMAL
);"""
)
```

```python
con.execute("INSERT INTO dim_produtos SELECT * FROM df2")
```

```python
df_dim_produtos = con.execute('SELECT * FROM dim_produtos').fetchdf()
df_dim_produtos.head()
```
